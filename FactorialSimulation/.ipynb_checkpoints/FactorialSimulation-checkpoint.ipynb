```julia
oneSidedResults = load("FactorialSimulationOneSided.jld")["results"]
twoSidedResults = load("FactorialSimulationTwoSided.jld")["results"]
```

```julia
#Mean over all but one attribute

function meanOverAllButOneIndex(index::Integer, arr::AbstractArray)
    averageOverIndices = (setdiff(2:length(size(arr)), index)...,)
    meanArray = mean(arr, dims=averageOverIndices)
    reshaped = dropdims(meanArray, dims=averageOverIndices)
    return reshaped
end


oneSidedNumClusters     = meanOverAllButOneIndex(2, oneSidedResults)
oneSidedNumObservations = meanOverAllButOneIndex(3, oneSidedResults)
oneSidedImbalance       = meanOverAllButOneIndex(4, oneSidedResults)
oneSidedPrecision       = meanOverAllButOneIndex(5, oneSidedResults)
oneSidedRandomizeRate   = meanOverAllButOneIndex(6, oneSidedResults)

twoSidedNumClusters     = meanOverAllButOneIndex(1, twoSidedResults)
twoSidedNumObservations = meanOverAllButOneIndex(2, twoSidedResults)
twoSidedImbalance       = meanOverAllButOneIndex(3, twoSidedResults)
twoSidedPrecision       = meanOverAllButOneIndex(4, twoSidedResults)
twoSidedRandomizeRate   = meanOverAllButOneIndex(5, twoSidedResults)
```

```julia
plot(oneSidedNumClusters[1, :], label="Fit", linewidth=2, linecolor=:green)
plot!(oneSidedNumClusters[2, :], label="sym", linewidth=2, linecolor=:blue)
plot!(oneSidedNumClusters[3, :], label="Flat", linewidth=2, linecolor=:red)
plot!(oneSidedNumClusters[4, :], label="perm", linewidth=2, linecolor=:black)
```

```julia
plot(twoSidedNumClusters[1, :], label="Fit", linewidth=2, linecolor=:green)
plot!(twoSidedNumClusters[2, :], label="sym", linewidth=2, linecolor=:blue)
plot!(twoSidedNumClusters[3, :], label="Flat", linewidth=2, linecolor=:red)
plot!(twoSidedNumClusters[4, :], label="perm", linewidth=2, linecolor=:black)
```
